```python
# Exercise 2: Get total profit of all months and show line plot with the following Style properties
''''
Generated line plot must include following Style properties: –
● Line Style dotted and Line-color should be red
● Show legend at the lower right location.
● X label name = Month Number
● Y label name = Sold units number
● Add a circle marker.
● Line marker color as read
● Line width should be 3
''''

df = pd.read_csv('company_sales_data.csv')

# Plotting the total profit over months with specified style properties
plt.plot(df['month_number'], df['total_profit'], marker='o', linestyle=':', color='red', markersize=8, linewidth=3, label='Total Profit')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')

# Adding legend at the lower right location
plt.legend(loc='lower right')

# Displaying the plot
plt.grid(True)
plt.show()
```

```python
#--------------------------EXERCISE 3:--------------------------------

# Extracting product names
products = df.columns[1:7]  # Assuming product columns start from the second column

# Plotting the number of units sold per month for each product
for product in products:
    plt.plot(df['month_number'], df[product], marker='o', label=product)

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Units Sold Per Month for Each Product')

# Adding legend
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()
```

```python
#--------------Exercise 4:----------------------------------------

# Extract toothpaste sales data for each month
toothpaste_sales = df['toothpaste']

# Generate x values (month numbers)
months = df['month_number']

# Plotting the toothpaste sales data using a scatter plot
plt.scatter(months, toothpaste_sales)

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Data')

# Adding grid with specified style
plt.grid(True, linestyle='--')

# Displaying the plot
plt.show()
```

```python
#---------------------------exercise 5:------------------------------

# Extract face cream and facewash sales data for each month
face_cream_sales = df['facecream']
face_wash_sales = df['facewash']

# Generate x values (month numbers)
months = df['month_number']

# Plotting the face cream and facewash sales data using a bar chart
plt.bar(months - 0.2, face_cream_sales, width=0.4, label='Face Cream')
plt.bar(months + 0.2, face_wash_sales, width=0.4, label='Face Wash')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Face Cream and Face Wash Sales Data')


# Adding legend
plt.legend()


# Adding grid with specified style
plt.grid(True, linestyle='--')

# Displaying the plot
plt.show()
```

```python
''''
EXERCISE 7:
    Read the total profit of each month and
show it using the histogram to see the most common
profit ranges
'''
# Extract total profit data for each month
total_profit = df['total_profit']

# Plotting the histogram of total profit
plt.hist(total_profit, bins=10, color='purple', edgecolor='purple')

# Adding labels and title
plt.xlabel(' Profit range in dollars')
plt.ylabel('Actual profit in dollars')
plt.title(' Profit data')


# Displaying the plot
plt.show()
```

```python
'''
EXERCISE 8: Calculate total sale data for last year for
each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.
'''
# Filter data for the last year (assuming the last 12 months)
last_year_data = df.tail(12)

# Calculate total sales for each product in the last year
total_sales_last_year = last_year_data.sum()

# Extract product names and corresponding total sales
product_names = total_sales_last_year.index[1:-2]  # Exclude non-product columns
sales_data = total_sales_last_year.values[1:-2]  # Exclude non-product values

# Plotting the pie chart
plt.figure(figsize=(8, 6))  # Set the figure size
plt.pie(sales_data, labels=product_names, autopct='%1.1f%%', startangle=140)
patches, texts, _ = plt.pie(sales_data, labels=product_names, autopct='%1.1f%%', startangle=140)
# Adding legend
plt.legend(patches, product_names, loc="best")

# Adding title
plt.title('Total Sales for Last Year by Product')

# Displaying the plot
plt.show()
```

```python
'''
Exercise 9: Read Bathing soap facewash of all months
and display it using the Subplot
'''

# Create DataFrame from provided data
data = {
    'month_number': list(range(1, 13)),
    'facewash': [1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760],
    'bathingsoap': [9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400]
}
df = pd.DataFrame(data)

# Extract data for bathing soap and facewash for all months
bathing_soap_data = df['bathingsoap']
facewash_data = df['facewash']
months = df['month_number']

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Plot bathing soap data
axes[0].plot(months, bathing_soap_data, marker='o', color='b',linewidth=3)
axes[0].set_title('Bathing Soap Sales Over Months')
axes[0].set_xlabel('Months')
axes[0].set_ylabel('Sales')

# Plot facewash data
axes[1].plot(months, facewash_data, marker='o', color='r',linewidth=3)
axes[1].set_title('Facewash Sales Over Months')
axes[1].set_xlabel('Months')
axes[1].set_ylabel('Sales')

# Adjust layout
plt.tight_layout()

# Display the subplots
plt.show()

```

```python
'''
Exercise Question 10: Read all product sales data and
show it using the stack plot
'''
# Create DataFrame from provided data
data = {
    'month_number': list(range(1, 13)),
    'facewash': [2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900],
    'toothpaste': [1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760],
    'bathingsoap': [5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400],
    'shampoo': [9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400],
    'moisturizer': [1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800]
}
df = pd.DataFrame(data)

# Extract product sales data
months = df['month_number']
facewash_sales = df['facewash']
toothpaste_sales = df['toothpaste']
bathingsoap_sales = df['bathingsoap']
shampoo_sales = df['shampoo']
moisturizer_sales = df['moisturizer']

# Create stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(months, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales, labels=['Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Product Sales Over Months')
plt.legend(loc='upper left')
plt.tight_layout()

# Display the stack plot
plt.show()
```
